```python
from tinydb import where
import sympy
class ViscosityModel(Model):
    """
    Metallic liquid viscosity model.
    Reference: Władysław Gąsior, "Viscosity modeling of binary alloys: Comparative studies",
        Calphad, Volume 44, 2014, Pages 119-128, ISSN 0364-5916,
        http://dx.doi.org/10.1016/j.calphad.2013.10.007.
    """
    def build_phase(self, dbe):
        super(ViscosityModel, self).build_phase(dbe)
        self.viscosity = self.build_viscosity(dbe)
        self.fragility = self.build_fragility(dbe)

    def build_viscosity(self, dbe):
        if self.phase_name != 'LIQUID':
            raise ValueError('Viscosity is only defined for LIQUID phase')
        phase = dbe.phases[self.phase_name]
        param_search = dbe.search
        eta_param_query = (
            (where('phase_name') == phase.name) & \
            (where('parameter_type') == 'ETA') & \
            (where('constituent_array').test(self._array_validity))
        )
        eta = self.redlich_kister_sum(phase, param_search, eta_param_query)
        liquid_mod = Model(dbe, self.components, self.phase_name)
        del liquid_mod.models['ref']
        del liquid_mod.models['idmix']
        excess_entropy = liquid_mod.entropy
        self.excess_entropy = excess_entropy
        ks = 2
        result = eta * (1 - ks * excess_entropy / v.R)
        return result
    def build_fragility(self, dbe):
        phase = dbe.phases[self.phase_name]
        param_search = dbe.search
        tg_param_query = (
            (where('phase_name') == phase.name) & \
            (where('parameter_type') == 'TG') & \
            (where('constituent_array').test(self._array_validity))
        )
        Tg = self.redlich_kister_sum(phase, param_search, tg_param_query)
        self.Tg = Tg
        fragility = (sympy.log(self.viscosity) / sympy.log(10)).diff(v.T).xreplace({v.T: Tg}) * -Tg
        return fragility
    @property
    def viscous_activation_energy(self):
        return v.R * sympy.log(self.viscosity).diff(v.T) * -(v.T**2)
```

```python
import numpy as np
mod = ViscosityModel(dbf, ['AL', 'CU', 'ZR'], 'LIQUID')
x1 = np.linspace(0,1, num=100)
x2 = np.linspace(0,1, num=100)
mesh = np.meshgrid(x1, x2)
X = mesh[0]
Y = mesh[1]
mesh_arr = np.array(mesh)
mesh_arr = np.moveaxis(mesh_arr, 0, 2)
dep_col = 1 - np.sum(mesh_arr, axis=-1, keepdims=True)
mesh_arr = np.concatenate((mesh_arr, dep_col), axis=-1)
orig_shape = tuple(mesh_arr.shape[:-1])
mesh_arr = mesh_arr.reshape(-1, mesh_arr.shape[-1])
mesh_arr[np.any(mesh_arr < 0, axis=-1), :] = np.nan
res = calculate(dbf, ['AL', 'CU', 'ZR'], 'LIQUID', T=1600, P=101325,
                model=mod, output='viscosity', points=mesh_arr)
res_eta = res.viscosity.values.reshape(orig_shape)

res = calculate(dbf, ['AL', 'CU', 'ZR'], 'LIQUID', T=1600, P=101325,
                model=mod, output='fragility', points=mesh_arr)
res_fragility = res.fragility.values.reshape(orig_shape)
```
